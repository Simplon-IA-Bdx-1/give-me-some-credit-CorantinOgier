```python
from bigml.api import BigML
import numpy as np
from pandas import DataFrame, read_csv
import kaggle
import matplotlib.pyplot as plt
from pickle import load, dump
import pprint
pp = pprint.PrettyPrinter(indent=4)
from os import path
```

```python
# On se connecte à BigML

api = BigML(project = 'project/5d94a41242129f2e16000244')
```

```python
# On importe nos deux fichiers csv train et test, modifiés, dans la source BigML

source_train = api.create_source('GMSC_modif_train.csv')
source_test = api.create_source('GMSC_modif_test.csv')

# On créé notre dataset avec le train et le test

train_dataset = api.create_dataset(source_train, {"name": "GMSC - Dataset Train Full"})
test_dataset = api.create_dataset(source_test, {"name": "GMSC - Dataset Test"})

# On split notre dataset train full en 80-20

split_train = api.create_dataset(
    train_dataset, {"name": "GMSC - modif train 80 | Training", "sample_rate": 0.8, "seed": "my seed"})
api.ok(split_train)

dev_set = api.create_dataset(
    train_dataset, {"name": "GMSC - Dev set", "sample_rate": 0.8, "seed": "my seed", "out_of_bag": True})
api.ok(dev_set)
```

```python
# On utilise pickle jar pour pouvoir travailler avec plusieurs pickle

jar_filename = 'jar/GMSC-picklejar'

# Si le fichier existe on l'ouvre, s'il n'existe pas on crée alors un dictionnaire vide

if path.exists(jar_filename): 
    with open(jar_filename, 'rb') as file:
        project_data = load(file)
else:
    project_data = {}
    
pp.pprint(project_data)
```

```python
# On va créer une boucle de variation de 10 à 100% de notre train dataset

for i in range(1, 11):
    
# On crée le dictionnaire 

    if i not in project_data:
        project_data[i] = {}

# Puis le dataset

    if 'dataset' not in project_data[i]:
        project_data[i]['dataset'] = {}
        
# Les différents modèles

    if 'models' not in project_data[i]:
        project_data[i]['models'] = {}
    if 'Ensemble' not in project_data[i]['models']:
        project_data[i]['models']['Ensemble'] = {}
    if 'Deepnet' not in project_data[i]['models']:
        project_data[i]['models']['Deepnet'] = {}
        
# Les évaluations
    if 'evaluations' not in project_data[i]:
        project_data[i]['evaluations'] = {}

# Le dev set

    if 'dev_set' not in project_data[i]['evaluations']:
        project_data[i]['evaluations']['dev_set'] = {}
    if 'Ensemble' not in project_data[i]['evaluations']['val_set']:
        project_data[i]['evaluations']['val_set']['Ensemble'] = {}
    if 'Deepnet' not in project_data[i]['evaluations']['val_set']:
        project_data[i]['evaluations']['val_set']['Deepnet'] = {}

# Le Train set  

    if 'train_set' not in project_data[i]['evaluations']:
        project_data[i]['evaluations']['train_set'] = {}
    if 'Ensemble' not in project_data[i]['evaluations']['train_set']:
        project_data[i]['evaluations']['train_set']['Ensemble'] = {}
    if 'Deepnet' not in project_data[i]['evaluations']['train_set']:
        project_data[i]['evaluations']['train_set']['Deepnet'] = {}
        
# On charge notre dataset, sinon on le crée à partir de notre train 80
    
    if 'dataset_id' in project_data[i]['dataset']:
        split_train_2 = api.get_dataset(project_data[i]['datase']['dataset_id'])
        api.ok(split_train_2)
    else:
        split_train_2 = api.create_dataset(split_train, {"name": "GMSC - Split Train 2 " 
                                                         + str(i/10), "sample_rate": i/10, "seed": "my seed"})

# On charge nos modèles Ensemble, sinon on les crée
    
    if 'model_id' in project_data[i]['models']['Ensemble']:
        ensemble = api.get_ensemble(project_data[i]['models']['Ensemble']['model_id'])
        api.ok(ensemble) 
    else:
        ensemble = api.create_ensemble(slit_train_2, {"objective_field" : "SeriousDlqin2yrs", "name": "Ensemble " 
                                                      + str(i/10)})
        api.ok(ensemble)
        project_data[i]['models']['Ensemble']['model_id'] = ensemble['resource']
        
# On charge nos modèles Deepnet, sinon on les crée

    if 'model_id' in project_data[i]['models']['Deepnet']:
        deepnet = api.get_deepnet(project_data[i]['models']['Deepnet']['model_id'])
        api.ok(deepnet)
    else:
        deepnet = api.create_deepnet(split_train_2, {"objective_field" : "SeriousDlqin2yrs", "name": "Deepnet " 
                                                                 + str(i/10)})
        api.ok(deepnet)
        project_data[i]['models']['Deepnet']['model_id'] = deepnet['resource']
        
# Pareillement pour le dev set, premièrement pour nos modèles Ensemble

    if 'evaluation_id' in project_data[i]['evaluations']['dev_set']['Ensemble']:
        evaluation_devset_ensemble = api.get_evaluation(
            project_data[i]['evaluations']['dev_set']['Ensemble']['evaluation_id'])
        api.ok(evaluation_devset_ensemble)
        project_data[i]['evaluations']['dev_set']['Ensemble']['auc'] = evaluation_devset_ensemble['object']['result']['model']['average_area_under_roc_curve']
    else:
        evaluation_devset_ensemble = api.create_evaluation(ensemble, dev_set)
        api.ok(evaluation_devset_ensemble)  
        project_data[i]['evaluations']['dev_set']['Ensemble']['evaluation_id'] = evaluation_devset_ensemble['resource']
        project_data[i]['evaluations']['dev_set']['Ensemble']['auc'] = evaluation_devset_ensemble['object']['result']['model']['average_area_under_roc_curve']

# Dans un second temps, les modèles Deepnet
        
    if 'evaluation_id' in project_data[i]['evaluations']['dev_set']['Deepnet']:
        evaluation_devset_deepnet = api.get_evaluation(project_data[i]['evaluations']['dev_set']['Deepnet']['evaluation_id'])
        api.ok(evaluation_devset_deepnet)
        project_data[i]['evaluations']['dev_set']['Deepnet']['auc'] = evaluation_devset_deepnet['object']['result']['model']['average_area_under_roc_curve']
    else:
        evaluation_devset_deepnet = api.create_evaluation(deepnet, dev_set)
        api.ok(evaluation_devset_deepnet)  
        project_data[i]['evaluations']['dev_set']['Deepnet']['evaluation_id'] = evaluation_devset_deepnet['resource']
        project_data[i]['evaluations']['dev_set']['Deepnet']['auc'] = evaluation_devset_deepnet['object']['result']['model']['average_area_under_roc_curve']
        
# On va procéder de la même manière pour le train set
# Pour Ensemble

    if 'evaluation_id' in project_data[i]['evaluations']['train_set']['Ensemble']:
        evaluation_trainset_ensemble = api.get_evaluation(project_data[i]['evaluations']['train_set']['Ensemble']['evaluation_id'])
        api.ok(evaluation_trainset_ensemble)
        project_data[i]['evaluations']['train_set']['Ensemble']['auc'] = evaluation_trainset_ensemble['object']['result']['model']['average_area_under_roc_curve']
    else:
        evaluation_trainset_ensemble = api.create_evaluation(ensemble, split_train_2)
        api.ok(evaluation_trainset_ensemble)  
        project_data[i]['evaluations']['train_set']['Ensemble']['evaluation_id'] = evaluation_trainset_ensemble['resource']
        project_data[i]['evaluations']['train_set']['Ensemble']['auc'] = evaluation_trainset_ensemble['object']['result']['model']['average_area_under_roc_curve']
        
# Pour Deepnet
        
    if 'evaluation_id' in project_data[i]['evaluations']['train_set']['Deepnet']:
        evaluation_trainset_deepnet = api.get_evaluation(project_data[i]['evaluations']['train_set']['Deepnet']['evaluation_id'])
        api.ok(evaluation_trainset_deepnet)
        project_data[i]['evaluations']['train_set']['Deepnet']['auc'] = evaluation_trainset_deepnet['object']['result']['model']['average_area_under_roc_curve']
    else:
        evaluation_trainset_deepnet = api.create_evaluation(deepnet, split_train_2)
        api.ok(evaluation_trainset_deepnet)  
        project_data[i]['evaluations']['train_set']['Deepnet']['evaluation_id'] = evaluation_trainset_deepnet['resource']
        project_data[i]['evaluations']['train_set']['Deepnet']['auc'] = evaluation_trainset_deepnet['object']['result']['model']['average_area_under_roc_curve']
        
        
    print(f"Fin de la boucle {i}")
    i += 1
    
# On enregistre le dictionnaire project_data dans un fichier via la fonction save_to
with open(jar_filename, 'wb') as file:
    dump(project_data,file)

print('Terminé !')
        
        
```
