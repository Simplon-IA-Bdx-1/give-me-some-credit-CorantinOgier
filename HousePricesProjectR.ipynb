Etude des attributs rares et leur impact sur SalesPrice

````r
# Récupération des colonnes avec NA > 300

AttributRare<-train[,c(which(colSums(is.na(train))>300),82)]
head(AttributRare)
summary(AttributRare)

# La valeur NA dans ces colonnes porte une information importante : absence de la feature! Remplaçons par "none"

levels(AttributRare$Alley)<-c(levels(AttributRare$Alley),"none")
levels(AttributRare$FireplaceQu)<-c(levels(AttributRare$FireplaceQu),"none")
levels(AttributRare$PoolQC)<-c(levels(AttributRare$PoolQC),"none")
levels(AttributRare$Fence)<-c(levels(AttributRare$Fence),"none")
levels(AttributRare$MiscFeature)<-c(levels(AttributRare$MiscFeature),"none")
AttributRare[is.na(AttributRare)] <- "none"
````

Dans l'ACP, il est intéressant de regarder les points qui s'éloignent de l'origine.

On peut sélectionner les colonnes numériques par exemple de la manière suivante:


Sinon, on peut accéder aux types des différentes colonnes comme suit:

````r
lapply(TrainFull, class)
````

````r
# Exemple de Régression Linéaire simple en fonction du GarageArea % Saleprice

reg<-lm(SalePrice~GarageArea, data = TrainFull)
summary(reg)
````

````r
# Régression Linéaire Multiple

numsTrainFull<-dplyr::select_if(TrainFull, is.numeric) # Sélection colonnes numériques de TrainFull

reg.multiple<-lm(SalePrice~., data = numsTrainFull) # ~. toutes les colonnes

summary(reg.multiple)
````

On souhaite procéder à un VIF, cependant il y a des features dans notre numsTrainFull qui doivent être supprimés. On va faire un alias:

````r
alias(reg.multiple)
````

On va donc devoir supprimer:

- TotalBsmtSF (BsmtFinSF1 et BsmtFinSF2)
- GrLivArea (X1stFlrSF, X2ndFlrSF et LowQualFinSF)

Puis refaire notre regression multiple afin de pouvoir avoir notre VIF.

````r
numsTrainFull<-subset(numsTrainFull, select=-c(TotalBsmtSF, GrLivArea))
reg.multiple<-lm(SalePrice~., data = numsTrainFull)
vif(reg.multiple)
````

Nous avons vu que les features concernant les Garages sont importantes, on va devoir fixer un seuil qui exprimera la redondance de certaines features au-delà de notre seuil.
