# GMSC with sklearn

## Load data

```python
from pandas import read_csv
train_full = read_csv('GMSC_modif_train.csv', index_col=0)
```

## Fix seed

```python
from numpy import random
SEED = 42
random.seed(SEED)
```

## Split data

```python
from sklearn.model_selection import train_test_split
VAL_SIZE = 0.2
train, val = train_test_split(train_full, test_size=VAL_SIZE)
```

Sanity check on shapes :

```python
print("Train full shape: " + str(train_full.shape))
print("Train shape: " + str(train.shape))
print("Val shape: " + str(val.shape))
```

## Saving data

```python
train.to_csv("GMSC_sk_train.csv")
val.to_csv("GMSC_sk_val.csv")
```
